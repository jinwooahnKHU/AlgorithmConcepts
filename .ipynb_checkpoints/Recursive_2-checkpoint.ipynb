## 예제 1

1부터 num까지 곱을 출력하는 함수

```python
#1. 반복문

def multiple(num):
    return_value = 1
    for index in range(1, num + 1):
        return_value = return_value * index
    return return_value
```

```python
#재귀

def multiple(num):
    if num <=1 :
        return num
    else:
        return num * multiple(num - 1)
```

```python
multiple(10)
```

## 예제 2

숫자가 들어있는 리스트가 주어졌을 때, 리스트의 합을 리턴하는 함수

```python
import random

data = random.sample(range(100),10)

data
```

```python
def sum_list(data):
    if len(data) <= 1:
        return data[0]
    
    #맨 처음 데이터만 떼고 나머지를 다시 함수에 넣는다.
    return data[0] + sum_list(data[1:])
```

```python
sum_list(data)
```

## 예제 3

회문(Palindrome) 은 순서를 거꾸로 읽어도 제대로 읽은 것과 같은 단어와 문장을 의미함


    ex. level -> 바꿔도 똑같음. 

Palindrome 을 판별할 수 있는 함수를 list slicing을 활용해서 만들어라

s t r i n g

string[0] == string[-1] 이거 하고 제외한 나머지를 넣으면 됨

-> 이게 나머지네 string[1:-1]


```python
def palindrome(data):
    if len(data) <= 1:
        return True
    else:
        if data[0] == data[-1]:
            return palindrome(data[1:-1])
        return False
```

```python
data = 'level'
palindrome(data)
```

## 예제 4

1. 정수 n에 대해

2. n이 홀수이면 3 x (n + 1) 을 하고,

3. n이 짝수이면 n을 2로 나눈다.

4. 반복하여 n 이 결국 1이 될 때까지 2와 3의 과정을 반복한다.


이 과정을 출력하는 함수를 만들어라

```python
def make_n1(data):
    data = int(data)
    #어차피 매번 출력해야 하므로 함수를 호출할 때 print하도록 한다.
    print(data)
    if data == 1:
        return data
    else:
        if data % 2 != 0:
            return make_n1(3 * data + 1)
        else:
            return make_n1(data / 2)
    
```

```python
make_n1(3)
```

## 예제 5

문제 : 정수 4를 1, 2, 3 의 조합으로 나타내는 방법은 다음과 같이 총 7가지가 있음

1 + 1 + 1 + 1<br>
1 + 1 + 2<br>
1 + 2 + 1<br>
2 + 1 + 1<br>
2 + 2<br>
1 + 3<br>
3 + 1

정수 n이 입력으로 주어졌을 때, n을 1,2,3 의 합으로 나타낼 수 있는 방법의 수를 구하시오.



연습장에 작성하는 습관을 기르자

<b>이 그림을 통해 f(n) = f(n - 1) + f(n - 2) + f(n - 3) 임을 알 수 있다. (n >= 4)</b>

```python
def func(data):
    if data == 1:
        return 1
    elif data == 2:
        return 2
    elif data == 3:
        return 4
    else:
        return func(data - 1) + func(data - 2) + func(data - 3)
```

```python
func(5)
```
