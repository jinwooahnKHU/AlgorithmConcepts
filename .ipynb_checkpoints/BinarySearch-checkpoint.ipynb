# 이진 탐색이란?

수형이랑 크롤링할 때 사용했던 방법이 이진 탐색 방법이였다!!

+ 탐색할 자료를 둘로 나누어 해당 데이터가 있을 만한 곳을 탐색하는 방법

## 이진 탐색의 이해(순차 탐색과 비교하며 이해하기)

+ 데이터가 정렬되어있다는 가정 하에 순차 탐색보다 훨씬 탐색 속도가 빠르다

## 분할 정복 알고리즘과 이진 탐색

+ 분할 정복 알고리즘(Divide and Conquer)
    - Divide : 문제를 하나 또는 둘 이상으로 나눈다.
    - Conquer : 나눠진 문제가 충분히 작고, 해결이 가능하다면 해결하고, 그렇지 않다면 다시 나눈다.
    
    
+ 이진 탐색
    - Divide : 리스트를 두 개의 서브 리스트로 나눈다.
    - Conquer 
        - 검색할 숫자(search) > 중간 값 이면, 뒷 부분의 서브 리스트에서 검색할 숫자를 찾는다.
        - 검색할 숫자(search) < 중간 값 이면, 앞 부분의 서브 리스트에서 검색할 숫자를 찾는다.
        
        
    - 재귀용법을 사용해서 분할정복 알고리즘을 적용해 이진 탐색을 실시한다.

# 코드로 구현

```python
def binary_search(data, search):
    if len(data) == 1 and search == data[0]: #data가 한개이고 그 값이 내가 찾는 값이면 True를 return
        return True
    #data가 한 개인데, 내가 원하는 값이 없는 경우
    if len(data) == 1 and search != data[0]:
        return False
    #방어코드
    if len(data) == 0:
        return False
    
    #중앙 index
    medium = len(data) // 2
    #만약 딱 중앙 Index의 값이 내가 원하는 값이면
    if search == data[medium]:
        return True
    else:
        if search > data[medium]:
            return binary_search(data[medium:], search)
        else:
            return binary_search(data[:medium], search)
```

```python
import random

data_list = random.sample(range(100),10)

data_list
```

```python
#먼저 탐색전에 정렬해주기
data_list.sort()
```

```python
binary_search(data_list, 5) #True면 있는 것ㅠ
```

# 순차 탐색

## 순차탐색이란?

+ 탐색은 여러 데이터 중에서 원하는 데이터를 찾아내는 것을 의미 -> 원하는 것을 있냐 없냐, 가져오냐 마느냐..

+ 탐색의 대표적인 기법이 순차탐색

+ 데이터가 담겨있는 리스트를 앞에서부터 하나씩 비교해서 원하는 데이터를 찾는 방법

임의 리스트가 다음과 같이 rand_data_list로 있을 때, 원하는 데이터의 위치를 리턴하는 순차탐색 알고리즘 작성해보기

- 원하는 데이터가 리스트에 없을 경우 -1을 리턴

```python
#풀이

from random import *

#rand_data_list 생성
rand_data_list = list()
for num in range(10):
    rand_data_list.append(randint(1,100))
```

```python
def sequential(data_list, search_data):
    for index in range(len(data_list)):
        #index의 데이터가 찾는 데이터면 index을 return
        if data_list[index] == search_data:
            return index
        else:
            return -1
```

```python
sequential(rand_data_list, 2)
```
