# 대표적인 정렬 : 삽입 정렬 (Insertion Sort)

기준점에서 하나 앞의 데이터와 크기를 비교하고 

만약 기준데이터가 더 작다면 바로 바꾸지 않고, 하나 더 앞의 데이터와 비교한다. 반복하여 맨 앞의 index까지 가거나 자신보다 큰 수가 없으면

그때 바꾼다.<br>
<br>




<b>
실제 적용

기준 데이터를 따로 빼고 그 자리를 비워놓고 하나 앞의 데이터와 비교해서 앞이 더 크다면 빈 자리로 옮기고 

더 앞의 index와 비교한다. 만약 자기보다 작다면 아까 움직인 데이터의 빈 자리에 기준 데이터 값을 넣는다.</b>

+ 기준점은 처음 index 하나 뒤 부터 시작한다.

+ 기준점 앞과 비교한다. [기준점, 기준점 -1 ], [기준점, 기준점 -2] ....

+ 기준 데이터 앞이 기준데이터 보다 작다면 그 앞은 이미 정렬이 되어있는 것이기 때문에 더 비교할 필요없이 다음으로 넘어가면 된다.


[데이터 길이, 조건 체크, 턴] 으로 비교하면 된다.

2,1,1


3,2,2


4,3,3


턴의 횟수가 곧 기준데이터의 index 위치가 된다.

``` python
    for index in range(데이터 길이 - 1):
        #index는 0부터 시작을 하니까 턴 시작은 1부터.. 따라서 index + 1 이고 거기서 1까지  -1 씩 변하면서 비교
        #range(10,0,-1) => 10 ~ 1까지 나옴
        for index2 in range(index + 1 ,0 , -1 ):
            if data[index2] < data[index2 -1 ]:
                swap(data[index2], data[index2 -1 ])
            #작지 않다면
            else:
                #더 앞을 비교할 필요는 없으니 
                break
```
               

## 구현

````python
def insertion_sort(data):
    #마지막은 비교할 필요없으니 len(Data) - 1
    for index in range(len(data) - 1):
        #기준점 하나 뒤부터 1까지 앞과 비교
        for index2 in range(index + 1,0,-1):
            #기준점보다 기준점 앞의 데이터가 크다면 swap
            if data[index2] < data[index2 - 1]:
                data[index2], data[index2 - 1]= data[index2 - 1],data[index2]
            else:
                break
    return data
````

````python
import random

data_list = random.sample(range(100),50)

print(insertion_sort(data_list))
````

## 시간 복잡도

for 문이 2개이므로 n 안에서 n 만큼 반복 한다.

반복문이 2개이므로 $O\left( n^{2} \right)$
