# 재귀 용법

+ 함수 안에서 동일한 함수를 호출하는 형태

+ 여러 알고리즘 작성시 사용되므로, 익숙해져야 함

+ 작성하는 패턴이 존재

## 예제 

팩토리얼

+ 간단한 경우부터 생각해보기
    - 2! = 1 x 2
    - 3! = 1 x 2 x 3
    - 4! = 1 x 2 x 3 x 4

+ 규칙이 보임 : n! = n x (n-1)!
    1. 함수를 하나 만든다.
    2. 함수(n)은 n > 1 이면 return n x 함수(n-1)
    3. 함수(n)은 n = 1이면 return n
    
+ 검증(코드로 검증하지 않고, 직접 간단한 경우부터 대입해서 검증해야 함)
    1. 먼저 2! 부터
        - 함수(2) 이면, 2 > 1 이므로 2 * 함수(1)
        - 함수(1) 은 1 이므로, return 2 x 1 = 2 맞다.
   
    2. 3! 
        - 함수(3) 이면, 3 > 1 이므로 3 * 함수(2)
        - 함수(2)은 2! 이므로 3 x 2! = 6 맞다ㅏ.

````python
def factorial(num):
    if num > 1:
        return num * factorial(num - 1)
    else:
        return num
````

````python
for num in range(10):
    print(factorial(num))
````

## 팩토리얼의 시간, 공간 복잡도

+ factorial(n)은 n-1 번의 factorial() 함수를 호출해서, 곱셈을 함
    - 일종의 n-1번 반복문을 호출한 것과 동일
    - factorial() 함수를 호출할 때마다, 지역변수 n이 생성됨
    
    
+ 시간복잡도/공간 복잡도는 O(n-1) 이므로 상수 필요없으니, 결국 둘 다 O(n)

## 재귀 호출의 일반적인 형태(패턴)

<b>일반적인 형태 1</b>
``` python
    def function(입력):
        if 입력 > 일정 값: #입력이 일정 값 이상이면 
            return function(입력 - 1) #입력보다 작은 값
        
        #결과적으로 반복하면 입력이 일정 값보다 작아지는 시점이 생기고 이 경우 아래 코드 실행
        else:
            return 일정 값, 또는 입력값 혹은 특정한 값 #재귀 호출 종료
```



<b>일반적인 형태 2</b>

위의 형태와 반대라고 생각해도 됨
``` python
    def function(입력):
        if 입력 <= 일정 값: #입력이 일정 값 이상이면 
            return 일정값 또는 입력 값
        #위의 if 문을 통과했다는 것은 입력이 일정값보다 크다는 것 그 경우에는 자기 함수를 호출
        function(입력보다 작은 값)
        return 결과 값
```

````python
#위에서 팩토리얼을 일반적인 형태 1 형식으로 구현했다면 

#아래는 일반적인 형태 2 방식으로 구현한 것

def factorial(num):
    if num <= 1:
        return num
    return_value = num * factorial(num - 1)
    return return_value
````

````python
for num in range(10):
    print(factorial(num))
````

<b>따라서, 입력값이 일정 값보다 클때, 작을때를 구분해서 그에 맞는 재귀함수를 구현해주면 된다!! </b>
