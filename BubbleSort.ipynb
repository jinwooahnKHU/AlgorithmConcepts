## 버블 정렬(bubble sort)이란?

+ 두 <b>인접한 데이터</b>를 비교해서, 앞에 있는 데이터가 뒤에 있는 데이터보다 크면, 자리를 바꾸는 정렬 알고리즘


<span style='font-size:15pt'>
<b> Idea </b>


1. 한 번 턴을 돌면, 가장 큰 값이 마지막 index에 들어가는 것을 알 수 있다.</span>
        - 그렇다면 그 다음 턴을 비교할 때, 굳이 마지막 index를 비교할 필요가 없다!! 
    
            왜냐하면 어차피 max값은 첫 턴에 가장 마지막으로 박힐 것이기 때문
            
            
2. 한번의 턴만에 모든 값이 정렬이 되었다. => 이후 턴으로 넘어갈 필요가 없다는 것은 어떻게 알 수 있을까?

    - 다음 턴에서 스왑이 한번도 일어나지 않았다면 더 이상 정렬이 필요하지 않다는 것을 의미한다.(early stop condition)

## 어떻게 코드로 만들까?

[데이터 길이, 조건체크, 턴] 이 경우의 수를 분석

조건체크 : 데이터 길이 - 1

턴 : 데이터 길이 - 1


패턴화하자면, 

``` python
    #턴을 표현
    for index in range(데이터 길이 - 1):
        #턴 안에서 조건 체크
        for index2 in range(데이터 길이 - 1):
            if 앞 데이터 > 뒤 데이터:
                swap(앞 데이터, 뒤 데이터)
```

### 데이터가 두 개일 때 버블 정렬 알고리즘 적용

+ 1 번 비교하면 됨.

### 데이터가 세 개일 때 버블 정렬 알고리즘 적용

+ index 0,1 을 비교 -> 1,2 비교 => 그러나 이렇게 해도 정렬은 되지 않는다. 

+ 다시 또 처음부터 비교해야 함. 즉, 한 바퀴 돌았다고 해서 한번에 정렬되는 것은 아님. 두 번 돌려야 함.


<b>
한 턴에 : 데이터길이  - 1 번 비교 하고, 

총 턴 하는 횟수는 데이터 길이 - 1 번 한다.


## 코드로 구현

````python
def bubblesort(data):
    #총 턴 수
    for index in range(len(data) - 1):
        #한 턴안에서 비교하는 횟수
        #한 턴을 할 때마다 최대값이 맨 마지막 inde로 가므로 마지막은 비교할 필요가 없음. 따라서 len(data) - index -1
        swap = False #swap이 일어났다는 것을 알기 위해
        for index2 in range(len(data) - index - 1):
            if data[index2] > data[index2 + 1]:
                data[index2], data[index2 + 1] = data[index2 + 1], data[index2]
                swap = True
        
        #한 턴안에서 스왑이 전혀일어나지 않았으면 정렬이 이미 된 상태이므로 더 이상 정렬으르 할 필요가 없으니 break
        if swap == False:
            break
    
    return data
            
````

Test

````python
import random

data_list = random.sample(range(100),50) #100개 중 50개를 뽑아서 넣는다.
bubblesort(data_list)
````

## 알고리즘 시간 복잡도

for 문이 2개이므로 n 안에서 n 만큼 반복 한다.

반복문이 2개이므로 $O\left( n^{2} \right)$ 



완전 정렬이 된 상태면 한번만 돌기 때문에 최선은 O(n)

최악의 경우 $\frac{n\left( n-1 \right)}{2}$
